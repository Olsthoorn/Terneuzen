## Inleiding

In dit document wordt de werkzaamheid onderzocht van een aantal varianten voor de bemaling van de sluishoofden voor de Nieuwe Sluis te Terneuzen die vanaf eind 2018 zal worden gebouwd.

Het binnen- en het buitenhoofd van de Nieuwe Sluis vormen de onderdelen die de diepste bemaling vragen voor de periode (maanden) waar binnen de sluishoofden moet worden gebouwd. Hierna volgt een geïllustreerde beschrijving van de situatie.

# Bemaling met zandpalen of putten

De meest ideale bemaling zou een bemalingvlak zijn direct onder de Boomse Klei, waarin de stijghoogte tot NAP -11 m wordt verlaagd. De doorlatendheid van het glauconietzand zou er niet toe doen, alleen de weerstand van de kleilaag tussen dit zand en Toz2 zou dan nog van belang zijn voor het totale waterbezwaar. Deze situatie is echter niet realiseerbaar. De bemaling, of die nu in het glauconietzand of in Toz2 zou moeten plaatsvinden, moet langs de binnenwanden van het sluishoofd worden geprojecteerd om de bouw niet te hinderen en gedurende de bemalingsperiode toegang tot de putten te behouden. Door de geringe spreidingslengte van het glauconietzand, zoals vastgesteld bij pompproef 4, is het onmogelijk om met putten in het glauconietzand aan de binnenwand van de diepwanden rond het sluishoofd de vereiste verlaging midden onder het sluishoofd te realiseren.

Een situatie die in de buurt komt zou er een kunnen zijn waarin het gehele sluishoofdoppervlak en de onderliggende Boomse Klei met het glauconietzan worden doorboord met zandpalen, zodat de vereiste verlaging wel kan worden gerealiseerd. Deze zandpalen wateren dan naar boven af in de grindlaag op NAP -20 m, waarop zij worden aangesloten. Het gehele waterbezwaar van de grindlaag wordt hierbij vervolgens afgevoerd via drains die in deze laag van tevoren zijn aangebracht.

Deze situatie is in principe maakbaar. Bij voldoende zandpalen moet ook de vereiste verlaging tot NAP -11 m haalbaar zijn, aangezien de verlaging in de grindlaag tot NAP -20 of -22 m gaat, 10 m verval dus. Echter, de zandpalen kunnen niet worden bijgeregeld wanneer hun opbrengst tegenzit. Om te kunnen bepalen of de zandpalen goed werken moet tijdens de bemaling de stijghoogte op voldoende plekken onder de Boomse klei worden gemeten. Dit is wel mogelijk via bronnen en peilbuizen die langs de diepwanden worden geplaatst, maar het zal moeilijk tot praktisch onmogelijk zijn om de stijghoogte midden onder het sluishoofd te meten. Om deze reden is een bemaling met zandpalen tot in het glauconietzand af te raden, want een risico. Een ander, groter gevaar is opbarsten of omhoog drukken van de betonvloer onderin de bouwkuip. Dit aspect wordt verderop behandeld.

Zandpalen tot in Toz2, onder het glauconietzand, hebben geen last van kleine een spreidinglengte; onder het glauconietzand is de spreidingslengte juist heel groot, mogelijk kilometers. Hier kan dus wel worden gegarandeerd dat de vereiste verlaging onder het gehele sluishoofd wordt bereikt. Dit zal met putten het geval zijn, maar kan evengoed met zandpalen. Het voordeel van zandpalen is dat zij over de gehele bouwput kunnen worden verdeeld en dus per stuk een lage capaciteit behoeven zodat een diepte tot enkele meters in Toz2 voldoende is. Zo wordt maximaal profijt wordt getrokken van de van nature in Toz2 aanwezige verticale anisotropie en ook van de diepte van de diepwanden, die dit effect versterken. Er zal dus geld bespaard kunnen worden door deze anisotropie te bepalen met een geschikte pomproef, mogelijk een onder elk van beide sluishoofden.

De derde variant bemaling bestaat uit putten in Toz2, geplaatst langs de binnenwand van de diepwanden. Het aantal benodgide putten wordt bepaald door de haalbare capaciteit per put versus het af te voeren waterbezwaar in Toz2 waarbij ook op een aantal reserveputten moet worden gerekend. In elk geval zullen de filters van deze putten een zekere lengte moeten krijgen en daarmee onder de gedachte diepte van NAP -45 m van de diepwanden uitkomen. Deze diepwanden helpen dan niet mee om het waterbezwaar te remmen. Het kan daarom bij deze bemaling noodzakelijk zijn de diepwanden tot grotere diepte aan te brengen dan was voorzien.

In hoeverre deze a priori overwegingen kloppen zal nu worden nagegaan door middel van grondwatermodellering.

# Modelleringsaanpak

Het te bemalen sluishoofd wordt hier gemodelleerd als een cirkelvormige bouwput met een radius van `r = 65 m`. Bij deze straal komen oppervlak en omtrek goed overeen met die van de echte sluishoofden, die bij benadering vierkant zijn. Het model is dus axiaal-symmetrisch, wat impliceert dat de bodemopbouw in alle richtingen gelijk wordt genomen. Deze benadering is geen enkel bezwaar omdat de spreidingslengte binnen Toz2 zo groot is, dat  verschillen in bodemopbouw in verschillende richtingen op de schaal van het sluishoofd niet of nauwelijks relevant zijn. De voorwaarde hiervoor is dat de lagen in het gat in de Boomse Klei voldoende verticale weerstand geven. Uit pompproef 4 blijkt dat het geval. Het is ongetwijfeld zo, dat de weerstand van de opvulling van het gat in de Boomse Klei met fijne, marine zanden veel minder is dan van de Boomse Klei zelf. Het is dus van belang dat voor de Boomse Klei in het model een representatieve waarde wordt gebruikt voor de weerstand, die mede, of zelfs vooral wordt bepaald door die van de genoemde lagen in het gat. Verder kan de "effectieve" weerstand van de Boomse Klei in het model lager zijn dan die van de Boomse Klei zelf, als gevolg van doorsnijdingen van de Boomse Klei in uitgeschuurde diepe sleuven in de Westerschelde. Wat de verticale opbouw betreft het het axiaal-symmetrische model geen beperkingen.

In een axiaal-symmetrisch model is een onttrekking in de werkelijkheid geen put maar een cirkelvormige onttrekkingsring. De onttrekking over zo'n ring is veel gelijkmatiger, en de afpomping geringer, dan met afzonderlijke putten. Er moet bij een axiaal-symmetrisch model dus een aparte slag worden gemaakt die de onttrekkingsring vertaalt naar een te kiezen aantal putten. Dit kan op verschillende manieren.

De bemaling van het glauconietzand wordt aldus gesimuleerd door een ring van 0.5 m wijdte tegen de binnenwand van de diepwanden tussen NAP -36 en -42 m. De bemaling van Toz2 geschiedt in het model met een ring van 0.5 m wijdte tegen de binnenwand van de diepwanden van NAP -43 tot NAP -49 m, dat wil zeggen met 6 m omstortingslengte in Toz2, waarin een putfilter van 5 m lengte past. Uiteraard kunnen deze putten tegelijk in het glauconietzand en Toz2 van een filter worden voorzien, maar dat levert a) weinig op omdat het doorlaatvermogen van het glauconietzand verwaarloosbaar is ten opzichte van dat van Toz2, en b) bestaat het gevaar dat het water uit het glauconietzand veel meer deeltjes meevoert, wat het ongeschikt maakt voor eventuele voeding van retourputten.

De bemaling met zandpalen zal eveneens vanuit Toz2 geschieden omdat dan een veel grotere garantie bestaat dat zij daadwerkelijk zullen werken. Uiteraard zullen zandpalen ook water uit het glauconietzand afvoeren aangezien zij open zijn naar alle lagen waar zij doorheen zijn geboord. Modelmatig wordt een bemalingsvlak dat bezet is met zandpalen als een onttrekkingsschijf ingevoerd met radius gelijk aan die van de bouwput en geplaatst tot 2 m onder de dunne kleilaag aan de top van Toz2. Uiteraard moet ook in deze situatie een vertaling worden gemaakt van de verlaging in het modelmatige onttrekkingsvlak naar die van een groot aantal zandpalen. Het model berekent de benodigde afpomping door in het onttrekkingsvlak de gewenste stijghoogte van NAP -11 m op te leggen. Waarna deze vertaalt moet worden naar de stijghoogte in de zandpalen en in de lengterichting van de zandpalen.

# Het model

## Inlezen van de dikte van de lagen en hun doorlatendheden

De functie `get_params()`, hieronder gedefinieerd, leest de modelparameters uit het opgegeven excel bestand en geneert vervolgens de modellagen waarmee gerekend zal worden.

```python
def get_params(workbook=None, sheetname=None, header=0):
    '''Return model parameters from workbook/sheetname.

    parameters
    ----------
        workbook : Excel workbook
            workbook with parameters
        sheetname : sheetname in excel workbook
            sheet with the table with parameters.

            Columns required are:
                name, ztop, zbot, D, nsub, kh, kv, ss
        header : int
            zero-based rownumber in sheet with the column headers

    '''


    cols=['name', 'ztop', 'zbot', 'D', 'nsub', 'kh', 'kv', 'ss', 'color', 'alpha']
    layers = pd.read_excel(workbook, sheetname='model', header=4)
    layers = layers[cols]

    ilay = 0
    layer = dict()
    for i in range(len(layers)):
        layer[ilay] = {col :layers[col].iloc[i] for col in layers.columns}
        nsub = layer[ilay]['nsub']
        if nsub > 1:
            layer[ilay]['D'] = layer[ilay]['D'] / nsub
            layer[ilay]['zbot'] = layer[ilay]['ztop'] - layer[ilay]['D']
            layer[ilay]['nsub'] = 1
            ilay += 1
            for j in range(nsub - 1):
                layer[ilay] = layer[ilay-1].copy()
                layer[ilay]['ztop'] = layer[ilay-1]['zbot']
                layer[ilay]['zbot'] = layer[ilay]['ztop'] - layer[ilay]['D']
                ilay += 1
        else:
            ilay += 1
    return layer
```

### Diepwand en onttrekking
De onttrekking wordt gekenmerkt door zijn binnen- en buitenradius `rwell` en door zijn hoogte `zwell` en tenslotte door de daarin te handhaven stijghoogte `hwell` of het onttrekkingsdebiet. Stijghoogte en debiet zijn uitwisselbaar, bij gegeven stijghoogte berekent het model de onttrekking, en omgekeerd, bij gegeven onttrekking berekent het model de stijghoogte.
```python
zwand = (0., variant[iv]['zwand'])
rwell = (variant[iv]['rwell1'], variant[iv]['rwell2'])
zwell = (variant[iv]['zwell1'], variant[iv]['zwell2'])
```

## Het rekennetwerk en de doorlatendheden

De laagopbouw en de verticale en horizontale doorlatendheid van elke laag wordt ingelezen uit het Excel bestand `paramsGatBoomseKlei.xls`; `layer` is een array met de eigenschappen `z`, `kh`, `kv` en `ss` (specifieke bergingscoëfficiënt) van elke laag.

```python
#%% Grid en laagopbouw

z = [layer[0]['ztop']]  # top model is top van laag 0

# Haal de laageigenschappen uit de ingelezen lagen
kh = []
kv = []
ss = []
for i in range(len(layer)):
    z.append(layer[i]['zbot'])
    kh.append(layer[i]['kh'])
    kv.append(layer[i]['kv'])
    ss.append(layer[i]['ss'])

y  = [ -0.5, 0.5] # y coordinaat = dummy voor axiaal symmetrisch model

# Alle benodigde afstanden voor het rekennetwerk. Zij worden
# gesorteerd en dubbelingen worden automatisch verwijderd.
# Buitenradius van het model is 10e4, zie logspace(..)
r  = np.hstack((0, *rwell, *rwand, np.logspace(0, 4, 101)))
gr = Grid(r, y, z, axial=True)
```

# Mask-arrays, zones binnen het model

Mask arrays zijn arrays om model cellen in bepaalde zones direct te adresseren, zoals de diepwand, de onttrekking en de vloer van de bouwkuip. 

```python
# Berekenen van zones in het model om de cellen daarin direct
# te kunnen adressere. Wand is de zone van de cellen in de diepwand.
wand = np.zeros(gr.shape, dtype=bool)
wand[AND( AND( AND(gr.XM > rwand[0], gr.XM < rwand[1]),
                    gr.ZM < zwand[0]),
                    gr.ZM > zwand[1])] = True

# well is de zone van de cellen binnen de onttrekking
well = np.zeros(gr.shape, dtype=bool)
well[AND( AND( AND(gr.XM  > rwell[0], gr.XM < rwell[1]),
                    gr.ZM < zwell[0]),
                    gr.ZM > zwell[1])] = True

# vloer zijn de cellen van de grindlaag op de Boomse klei (niet nodig
# voor het model)
vloer = np.zeros(gr.shape, dtype=bool)
vloer[AND(gr.XM < rwand[0], gr.ZM > -21.9)] = True
```

## Bodemeigenschappen en randvoorwaarden

Deze worden eerst aangemaakt voor het hele model en vervolgens aangepast waar nodig zoals in de diepwand. Deze eigenschappen per geologische laag worden gelezen uit het Excel werkboek `paramsGatBoomseKlei.xls`.

```python
Kh = gr.const(np.array(kh));  Kh[wand] = kwand
Kv = gr.const(np.array(kv))  # verticale doorlatendheid
Ss = gr.const(np.array(ss))  # specifieke bergingscoefficient
```

## Randvoorwaarden

De IBOUND array geeft aan waar in het model de stijghoogte wordt opgegeven (-1) en waar het model niet actief is(0)

```python
#%% Boundary array and boundary conditions
IBOUND = gr.const(1, dtype=int)
IBOUND[0]    = -1
IBOUND[well] = -1
IBOUND[vloer] = 0

# Opgelegd debiet, 0 waar niets opgelegd
FQ = gr.const(0.)

# Initiële stijghoogte. Alleen van belang waar opgelegd, de andere worden berekend.
HI = gr.const(0.) # m NAP, dummy
HI[well] = hwell  # m NAP, opgelegd.
```

## Simulatietijd

Voor de bemaling is alleen de stationaire verlaging van belang, want die wordt gemakkelijk gehaald tijdens de tijd dat bemalen moet worden. Hier wordt echter niet-stationair gerekend tot de eindsituatie is bereikt. Dit geeft dus ook een idee over het niet-stationaire gedrag en de tijd waarin de eindsituatie zal worden bereikt.

```python
#%% simulation time
t = np.logspace(-3, 3, 101) # times for simulation
```

# Observatiepunten

Hieronder wordt een aantal observatiepunten gedefinieerd, waarvoor de tijdsafhankelijke verlaging zal worden getoond. Elk punt heeft een naam een radius (afstand tot het hart van het model) en het niveau in m NAP.

Het de plek van het punt 'well' hangt van de berekende variant af en dus geparametriseerd.

Putten 2, 3 en 4 liggen midden onder de bouwput (`r=1` m) op respectievelijk NAP -40, -50 en -70 m, dus in het glauconietzand, boven in Toz2 en in Toz3. De andere punten liggen in Toz2 op steeds grotere afstand. Het punt op 5000 m fungeert om de controleren in hoeverre de gesloten buitenrand van het model de uitskomsten beïnvloedt; wanneer de verlaging op dat punt verwaarloosbaar is, is het model ruim genoeg. De verlagingsgrafieken laten zien dat dit inderdaad het geval is.

```python
rw = np.round(0.5 * (rwell[0] + rwell[1]), 1)  # hart van de onttrekking
zw = np.round(0.5 * (zwell[0] + zwell[1]), 1)  # hart van de onttrekking

points = [('well_{:.0f}'.format(rw), rw, zw), # hart van de onttrekking
          ('P0001_40',    1., -40.),  # in glauconietzand, onder hoofd
          ('P0001_50',    1., -50.),  # in Toz2, midden onder sluishoofd
          ('P0001_70',    1., -70.),  # in Toz3, midden onder sluishoofd
          ('p0075_50',   75., -50.),  # in Toz2, op r = 75 m
          ('p0250_50',  250., -50.),  # in Toz2, op r = 250 m
          ('p0500_50',  500., -50.),  # in Toz2, op r = 500 m
          ('p1000_50', 1000., -50.),  # in Toz2, op r = 1000 m
          ('p5000_50', 5000., -50.)]  # in Toz2, op r = 5000 m
```

## Berekening van de verlagingen en stijghoogtes

Het eindige differentiemodel Fdm3t rekent de verlagingen en grondwaterstanden uit.

```python
fdmObj = Fdm3t(gr=gr, t=t, kxyz=(Kh, Kh, Kv), Ss=Ss, FQ=FQ, HI=HI,
            IBOUND=IBOUND, epsilon=1.0)
```

Sommeren over alle cellen van laag 1 levert de totale infiltratie op en sommeren over alle overige cellen levert de totale onttrekking op. Beide zijn aan elkaar gelijk met tegengesteld teken. We laten verschillende delen in de doorsnede zien met behulp van patches (hier gekleurde, soms transparente rechthoeken) waar de stijghoogtelijen en de stroomlijnen overheen getekend worden.

## Vertaling van onttrekkingsring naar indiviuele putten
Bij wandlengtes van de sluishoofden van 140 m ongeveer, is het redelijk om lang elke diepwand 5 putten te plaatsen, ca. 28 m onderlinge afstand. Dit komt dan neer op 16 putten rondon.

De verlaging een in put bij 16 putten op een cirkel ten opzichte van verlaging van een ring kan worden geschat bij een aanname van de afstand `R` waarop de verlaging nul is. In het onderhavige model is de verticale weerstand van alle lagen samen ca. 7000 d en het doorlaatvermogen ca. 100 m2/d. De spreidinglengte is dan bijna 1000 m. De verlaging op de rand van de bouwput in Toz2 en Toz3 kunnen we dan berekenen met de eenvoudige formule van Dupuit door voor de afstand waarop geen verlaging plaats vindt de spreidingslengte te gebruiken:
$$ s_r = \frac Q {2 \pi kD} \ln \left( \frac {1.123 \lambda} {r_b} \right) $$

waarin $r_b = 65$ m de radius van de bouwput is en $1.123 \lambda \approx 1000$ m

De verlaging in een put op een ring met $n$ putten is een kwestie van superpositie.

$$s_p = \sum_1^n \frac {Q / n} {2\pi kD} \ln \left( \frac {1.123 \lambda } {r_i} \right)$$

waarin $r_i$ de afstand van put nr $i$ tot de beschouwde put (die willekeurig is) en waarvoor de afstand gelijk is aan de putstraal $r_w$.



```python
n = 16 # aantal putten
r_b = rwand[0]

# coordinaten van putten op de ring
x = [r_b * np.cos(i/n * 2 * np.pi) for i in range(n)]
y = [r_b * np.sin(i/n * 2 * np.pi) for i in range(n)]

rw= 0.25 # putdiameter

# beschouwde punt, putwand met maximale verlaging
x0 = r_b
y0 = rw

# punt tussen twee putten met minimale verlaging
x1 = r_b * np.cos(0.5/n * 2 * np.pi)
y1 = r_b * np.sin(0.5/n * 2 * np.pi)

# Put debiet
Q = 60 * 24
kD = 72.     # Doorlaatvermogen Toz2 + Toz3
R0 = 1000.   # 1.123 * lambda

s0 = np.sum([-Q/ n /(2 * np.pi * kD) * np.log(np.sqrt((xx - x0)**2 + (yy-y0)**2) / R0) for xx, yy in zip(x, y)])
s1 = np.sum([-Q/ n /(2 * np.pi * kD) * np.log(np.sqrt((xx - x1)**2 + (yy-y1)**2) / R0) for xx, yy in zip(x, y)])
sr = Q /(2 * np.pi * kD) * np.log(R0 / r_b)

print('sr [ring      ]: {:.2f} m'.format(sr))
print('s0 [put       ]: {:.2f} m'.format(s0))
print('s1 [tussenpunt]: {:.2f} m'.format(s1))
```

De verlaging in de put is ongeveer 0.55 m ofwel 11% groter dan op de ring en in het tussenpunt is de verlaging ietje kleiner, zoals te verwachten is. De werkelijke verlaging van de ring is echter groter doordat nabij de putten het grootste deel van het water alleen via Toz2 toestroomt. In variant 1 is de verlaging van de onttrekkingsring 12.3 m. Ook hier kan 11% extra verlaging in de putten worden genomen, dus 1.4 m. Wellicht moet hier nog extra weerstand aan toevoegenten gevolge van de skin op de boorgatwand die bij het boren kan ontstaan en mogelijk niet geheel verwijder is. Een extra verlaging van 5 m lijkt dus zeer ruim genomen. Dit betekent dat met 16 putten rond de bouwput en een verlagin in de putten van ca. 17.5 m de bemaling zal moeten voldoen.

# Resultaten per variant

De tabel geeft een overizicht van de varianten die zijn doorgerekend en waarvoor ook de stijghoogte in de onttrekking is bepaald de leidt tot ene stijhoogte van NAP -11 m midden onder de bouwput (in het glauconietzand).

Een eerste conclusie is dat de invloed van de diepe laag Toz3 aanzienklijk zal zijn en niet verwaarloosd mag worden, tenzij is aangetoond dat de weerstand van de kleilaag tussen Toz2 en Toz3 groot is. Feit is dat we deze weerstand niet kennen en dus ook niet weten of de invloed van het diepere Toz3 mag worden verwaarloosd. Waarschijnlijk niet. Verwaarlozing leidt tot een 50% lager waterbezwaar in het model.

Bemaling in alleen het glauconietzand is niet mogelijk omdat de putten daarbij droog zouden worden gepompt zonder dat de vereiste stijghoogte onder het midden van de bouwput wordt gehaald. Variant 0 laat dat zien. De benodigde stijghoogteverlaging in de onttrekkingsring is ca. NAP -33 m. Bij bemaling met putten is echter een grotere verlaging nodig door de wandweerstand en omdat geen ring maar afzonderlijke putten moeten worden geraakt. Ergo de bemaling moet plaats vinden in Toz2. Variant 0 laat zien dat de benodigde stijghoogte in de onttrekkingsring NAP -12.2 m is en het bijbehorende onttrekkingsdebiet 67.8 m3/h (dus ca. 70 m3/h). Dat is niet eens zoveel meer dan bij in het theoretische geval van bemaling van alleen het glauconietzand (was ca. 54 m3/h). De reden is dat bij bemaling van het glauconietzand, door de geringe spreidingslengte, een veel grotere verlaging in de putten moet worden opgelegd om de verlaging van NAP -11 m in het midden van de bouwput te bereiken. In de Toz2 laag is de spreidingslengte zo'n honderd keer groter en is de stijghoogte die in de onttrekkingsring moet worden opgelegd NAP -12.2 m, dus nog slechts 1.2 m lager dan de gewenste stijghoogte van NAP -11 m onder het centrum van de bouwput. Uiteraard is de stijghoogte in de te gebruiken putten lager dan deze NAP -12.2 m. De vertaling moet hierna nog plaats vinden.

De volgende varianten laten de invloed zien van de diepte van de diepwand die stapsgewijs met 4 m worden verlengd in de volgende 3 varianten. Verlenging van de diepwanden van 45 tot 54 m reduceert het waterbezwaar van 68 to 51 m3/h, dus zeer beperkt. Verdere verlenging tot in de kleilaag die op NAP -59 m begint reduceert het watebezwaar to 16 m3/h. Let wel, bij verwaarlozing van de onderliggende lagen zou het waterbezwaar dan nul zijn. In het algemeen geldt, water remmen door verlengen van dam- of diepwanden is een dure, inefficiënte actie, tenzij de de wanden tot in een onderliggende kleilaag worden aangebracht en de lek door die kleilaag beperkt is.

Als alternatie zou met zandpalen kunnen worden gedraineerd. Ook deze moeten enkele, zeg 2 meters in Toz2 reiken om voldoende capaciteit te geven. Deze zandpalen draineren dan naar de grindlaag die op de Boomse Klei zal worden aangebracht. Het voordeel is dat zandpalen gemakkelijk over de hele oppevlak van de bouwkuip kunnen worden verspreid. De zandpalen werken dan gezamenlijk als een soort onttrekkingschijf. Theoretisch voordeel is dat de stijghoogte die in een onttrekkngsschijf moet worden opgelegd precies gelijk is aan de gewenste stijghoogte, dus NAP -11 m. Echter omdat de zandpalen puntonttrekkingen vormen is de genodigde stijghoogte die moet worden oplegd later dan NAP -11 m, maar waarschijnlijk niet veel lager. De vertaling moet nog worden gemaakt.

Zandpalen of putten in Toz, het maakt weinig uit voor het waterbezwaar. Het voordeel van de zandpalen (of feitelijk grindpalen) is dat zij op geen enkele manier de werkzaamheden nadelig beinvloeden. Een groot nadeel is dat de verlaging die zijn onder de Boomse Klei teweeg gaan brengen mogelijk veel te groot wordt, aangezien de stijghoogte in de grindlaag op de Boomse Klei tijdens de werkzaamheden in den droge op ca NAP -20 m zal liggen. En er is geen enkele mogelijkheid om de afvoer van de zandpalen onder de eenmaal aangebrachte vloer te remmen. Bovendien zal bij een te grote afvoer door de zandpalen niet alleen het waterbezwaar nodeloos groot zijn, maar ook zal de betonnen vloer omhoog woden gedrukt, wat uiteraard voorkomen moet worden. De conclusie is dus dat zandpalen geen goede oplossing bieden.

Als eindconclusie is nagenoeg onvermijdelijk dat de enige werkbare manier van bemalen van de sluishoofden is met behulp van putten geplaatst aan de binnenzijde (en eventueel aan de buitenzijde) van de diepwanden met filters in Toz2. Geschat wordt dat putfilters van 5 m in een omstorting die tot 6 m in Toz2 reikt voldoende zou kunnen zijn. Maar de filterlengte maakt nu feitelijk niet meer uit voor het waterbezwaar. Bij volkomen filters in Toz2 is het waterbezwaar 67.1 m3/h bij bij korte filters is dit ook zo 67.7 m3/h (het verschil tussen deze getallen komt omdat de opgelegde stijghoogte in de onttrekkingring iteratief bepaald is en niet helemaal exact is uitgevoerd. Kortom beide situaties, volkomen of korte filters kennen een waterbezwaar van ca. 70 m3/h. Dit geldt bij diepwanden tot NAP -45 m. Verlengen van de diepwanden met 4 m helpt niets. Verlengen met 8 m scheelt iets meer dan 20%. Verlengen van de diepwanden met 12 m tot in de klei scheelt 75% van het waterbezwaar.

Rekening houdend met het feit dat de bemaling met aparte putten plaats moet vinden en met een redelijk aantal van 16 putten langs de omtrek van de bouwput (ca. 30 m onderlinge afstand) en rekening houdend met een skinweerstand van 2.5 m, moet met 16 putten met een afpomping van 17.5 m de bemaling voldoen.
